# **MySQL Workbench에서 데이터베이스 생성하기**

## **1. 워크벤치 접속 및 기본 개념**
- MySQL에 접속 후, **워크벤치(Workbench)** 실행  
- 좌측 **SCHEMAS** 영역에서 데이터베이스 확인 가능  
- MySQL에서 **Schema = Database**  



## **2. 데이터베이스 생성(SQL문 작성 및 실행)**
- **쿼리 창(Query Editor)** 에서 SQL문 작성  
- 데이터베이스 생성 명령어:

  ```sql
  CREATE DATABASE copang_main;
  ```



-	번개 버튼(Execute) 클릭하여 실행



3. 데이터베이스 확인
-	SCHEMAS 영역은 자동으로 갱신되지 않음 → 새로고침(Refresh) 버튼 클릭
-	생성된 copang_main 데이터베이스가 목록에 추가됨



4. 핵심 포인트
-	SQL 문법은 명확한 단어로 구성되어 있어 쉽게 해석 가능
-	데이터베이스 생성은 CREATE DATABASE [DB명] 형식
-	실행 후 SCHEMAS에서 확인 가능하며, 새로고침 필요
